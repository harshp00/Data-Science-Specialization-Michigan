```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
bayern = pd.read_csv('../data_folder/bayern.csv')
bayern = bayern.drop(columns = ['Division', 'Avg.Att.[2]', 'Cup', 'Europe', 'Res.', 'Top Goalscorer(s)'])
bayern['Year'] = bayern['Season'].str[:4]
bayern['Year'] = bayern['Year'].astype('float')
bayern = bayern[bayern['Year'] > 1964]
bayern
```

```python
bayern['GpGame'] = bayern['Goals']/bayern['P']
bayern['WinProb'] = bayern['W']/bayern['P']
bayern
```

```python
dortmund = pd.read_csv('../data_folder/dortmund.csv')
dortmund = dortmund.drop(columns = ['Division', 'German Cup', 'Europe/Other', 'Res', 'Top goalscorer(s)'])
dortmund['Year'] = dortmund['Season'].str[:4].astype('int')
dortmund = dortmund[dortmund['Year'] > 1964]
dortmund['P'] = dortmund['P'].str.replace('\[.*\]', '')
dortmund['P'] = dortmund['P'].astype('float64')
dortmund
```

```python

dortmund['GpGame'] = dortmund['Goals']/dortmund['P']
dortmund['WinProb'] = dortmund['W']/dortmund['P']
dortmund
```

```python
plt.figure()
ax1 = plt.subplot(2, 1, 1)
plt.plot(bayern['Year'], bayern['GpGame'], color = '#F75D59', label = 'Bayern Goals/Game')
plt.plot(dortmund['Year'], dortmund['GpGame'], color = '#EAC117', label = 'Dortmund Goals/Game')
leg = plt.legend(framealpha = 0, loc = 9)
for text in leg.get_texts():
    plt.setp(text, color = 'gray', fontsize = 7)

ax1 = plt.gca()
ax1.set_ylabel('Goals per Game', color = 'gray')
ax1.set_title('Bayern vs Dortmund Goals/Game over 1965-2019', fontsize = 10, color = 'gray')
for spine in ax1.spines.values():
    spine.set_visible(False)
ax1.set_xticklabels([])

ax2 = plt.subplot(2, 1, 2)
plt.plot(bayern['Year'], bayern['WinProb'], color = '#F75D59', linestyle = '--', label = 'Bayern Win Prob')
plt.plot(dortmund['Year'], dortmund['WinProb'], color = '#EAC117', linestyle = '--', label = 'Dortmund Win Prob')
for spine in ax2.spines.values():
    spine.set_visible(False)
leg = plt.legend(framealpha = 0, loc = 9)
for text in leg.get_texts():
    plt.setp(text, color = 'gray', fontsize = 7)

ax2 = plt.gca()
ax2.set_xlabel('Year', color = 'gray')
ax2.set_ylabel('Winning Probability', color = 'gray')
ax1.set_title('Bayern vs Dortmund Win Prob over 1965-2019', fontsize = 10, color = 'gray')
#plt.show()
plt.savefig('../A4Figure.pdf')
```
